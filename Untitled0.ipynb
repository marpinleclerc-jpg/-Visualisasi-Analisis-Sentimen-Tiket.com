```python
# Import semua kebutuhan
!pip install google-play-scraper
import pandas as ppk
from transformers import pipeline
from google_play_scraper import Sort, reviews
import matplotlib.pyplot as plt
import seaborn as sns

# Load Model AI
print("Loading model...")
baksoayam = pipeline("sentiment-analysis", model="mdhugol/indonesia-bert-sentiment-classification")
print("Model siap digunakan!")
```

```python
joy = int(input("Mau ambil berapa review? "))
app_id = 'com.tiket.gits'

print(f"Mengambil {joy} review dari tiket.com...")
reslut, _ = reviews(
    app_id,
    lang='id',
    country='id',
    sort=Sort.NEWEST,
    count=joy
)

# Masukkan ke DataFrame
kt = ppk.DataFrame(reslut)
teks_review = kt['content'].tolist()
print("Data berhasil diambil!")
```

```python
kamus = {
    'LABEL_0' : 'positive',
    'LABEL_1' : 'neutral',
    'LABEL_2' : 'negative'
}

print("Memulai analisa sentimen...")
hasil_ai = baksoayam(teks_review)

# Siapkan list kosong
list_sentimen = []

for i, data in enumerate(hasil_ai):
    label_asli = data['label']
    skor = data['score']
    label_manusia = kamus.get(label_asli, "Tidak Diketahui")

    # Simpan hasil
    list_sentimen.append(label_manusia)

    print(f"Review ke-{i+1}: {teks_review[i]}")
    print(f"Analisis: {label_manusia} (Keyakinan: {skor*100:.2f}%)\n")

# Masukkan hasil prediksi ke DataFrame untuk visualisasi nanti
kt['Sentimen_Prediksi'] = list_sentimen
print("Analisa selesai!")
```

```python
hitung_sentimen = kt['Sentimen_Prediksi'].value_counts()

warna = {'positive': '#28a745', 'neutral': '#6c757d', 'negative': '#dc3545'}
warna_grafik = [warna.get(x, '#333333') for x in hitung_sentimen.index]

fig, axes = plt.subplots(1, 2, figsize=(14, 6))

axes[0].pie(hitung_sentimen, labels=hitung_sentimen.index, autopct='%1.1f%%',
            startangle=140, colors=warna_grafik, explode=[0.05]*len(hitung_sentimen))
axes[0].set_title(f'Persentase Sentimen Tiket.com\n(Total {len(kt)} Review)')

sns.barplot(x=hitung_sentimen.index, y=hitung_sentimen.values, ax=axes[1], palette=warna)
axes[1].set_title('Jumlah Review per Sentimen')
axes[1].set_xlabel('Kategori Sentimen')
axes[1].set_ylabel('Jumlah Review')

# Tampilkan
plt.tight_layout()
plt.show()
```
